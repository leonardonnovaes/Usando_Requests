```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('dados/linguagens_amazon.csv')
df2 = pd.read_csv('dados/linguagens_netflix.csv')
df3 = pd.read_csv('dados/linguagens_spotify.csv')
df4 = pd.read_csv('dados/linguagens_accenture.csv')
df5 = pd.read_csv('dados/linguagens_facebook.csv')
df6 = pd.read_csv('dados/linguagens_ibm.csv')
df7 = pd.read_csv('dados/linguagens_oracle.csv')
df8 = pd.read_csv('dados/linguagens_totvs.csv')
```

```python
df1 = df1.dropna()
df2 = df1.dropna()
df3 = df1.dropna()
df4 = df1.dropna()
df5 = df1.dropna()
df6 = df1.dropna()
df7 = df1.dropna()
df8 = df1.dropna()

print(df1.isnull().sum())
print(df2.isnull().sum())
print(df3.isnull().sum())
print(df4.isnull().sum())
print(df5.isnull().sum())
print(df6.isnull().sum())
print(df7.isnull().sum())
print(df8.isnull().sum())
```

```python
df_main = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
```

```python
df_main['linguagens'].value_counts()
```

```python
top10 = df_main['linguagens'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top10.index,
    y=top10.values
)

plt.xticks(rotation=90)
plt.title("Top 10 Linguagens Mais Utilizadas no mercado por projeto")
plt.xlabel("Linguagens")
plt.ylabel("Quantidade")

plt.show()
```
