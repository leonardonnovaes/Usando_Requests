```python
import requests

access_token = 'seu_token'
username = 'amzn'
url = f"https://api.github.com/users/{username}/followers"
headers = {'Authorization': 'Bearer ' + access_token,
           'X-GitHub-Api-Version': '2022-11-28'}
```

```python
followers_list = []
page_num = 1
while True:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        followers = response.json()
        if len(followers)==0:
            break
        followers_list.append(followers)
        page_num += 1
followers_list
```

```python
names_followers = []
for page in followers_list:
    for follower in page:
        names_followers.append(follower['login'])
names_followers
```

```python
len(names_followers)
```

```python
import pandas as pd

followers_amzn = pd.DataFrame()
followers_amzn['name_followers'] = names_followers
```

```python
followers_amzn.to_csv('amazon_followers.csv')
```
