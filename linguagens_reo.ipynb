# Obtendo dados dos repositórios

```python
headers = {'X-Github-Api-Version': '2022-11-28'}
```

```python
api_base_url = 'https://api.github.com'
owner = 'IBM'
url = f'{api_base_url}/users/{owner}/repos'
```

```python
response = requests.get(url, headers=headers)
response.status_code
```

```python
response.json()
```

```python
len(response.json())
```

```python
access_token = 'seutoken'
headers = {'Authorization': 'Bearer' + access_token,
           'X-Github-Api-Version': '2022-11-28'}
```

# Paginando os repos

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
len(repos_list)
```

```python
len(repos_list[0])
```

# Transformando os dados

```python
repos_list[0][2]['name']
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
repos_name
```

```python
len(repos_name)
```

```python
repos_list[1][1]['language']
```

```python
repos_language = []
for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
repos_language
```

# Criando o data frame

```python
import pandas as pd
```

```python
dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name
dados_amz['lenguage'] = repos_language
```

```python
dados_amz
```

```python
dados_amz.to_csv('amazon_languages.csv')
```

# Criando repositorio com POST

```python
api_base_url = 'https://api.github.com'
url = f'{api_base_url}/user/repos'

url
```

```python
access_token = 'seu_token'
headers = {
    'Authorization': 'Bearer ' + access_token,  # espaço depois de Bearer
    'X-GitHub-Api-Version': '2022-11-28'
}

data = {
    'name': 'linguagens-utilizadas',
    'description': 'Repositorio com as linguagens de prog da Amazon',
    'private': False
}

response = requests.post(url, json=data, headers=headers)
response.status_code
```

# Formato de arquivo

```python
import base64
```

```python
with open('amazon_languages.csv', 'rb') as file:
    file_content = file.read()

encoded_content = base64.b64encode(file_content)
```

```python
api_base_url = 'https://api.github.com'
username = 'leonardonnovaes'
repo = 'linguagens-utilizadas'
path = 'amazon.csv'

url = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'
url
```

```python
data = {
    'message': 'Adicionando arquivo',
    'content': encoded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
response.status_code
```
